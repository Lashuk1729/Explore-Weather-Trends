```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('results.csv')
data.info()
```

```python
data.mean()
data['local_avg_temp'].fillna((data['local_avg_temp'].mean()), inplace=True)
data.info()
```

```python
data.describe()
```

```python
sns.boxplot(data['global_avg_temp'], color = 'orange')
sns.swarmplot(data['global_avg_temp'], color = '.35')
```

```python
sns.boxplot(data['local_avg_temp'], color = 'yellow')
sns.swarmplot(data['local_avg_temp'], color = '.35')
```

```python
def moving_avg(data_series,win):
    out = data_series.rolling(window = win, center = False).mean()
    return out
```

```python
plt.figure(figsize=(15, 6))
plt.plot(data['year'], data['global_avg_temp'], label = 'Actual Global Temp.')
plt.legend()
plt.xlabel("Years")
plt.ylabel("Temperature (C)")
plt.title("Global Average Temperature")
plt.show()
```

```python
plt.figure(figsize=(15, 8))
plt.plot(data['year'], moving_avg(data['global_avg_temp'], 5), label='5 per. Moving Global Temp.', alpha = 0.6)
plt.plot(data['year'], moving_avg(data['global_avg_temp'], 10), label='10 per. Moving Global Temp.',alpha = 0.9)
plt.plot(data['year'], moving_avg(data['global_avg_temp'], 50), label='50 per. Moving Global Temp.')
plt.plot(data['year'], data['global_avg_temp'], label = 'Actual Global Temp.',alpha = 0.3)
plt.legend()
plt.xlabel("Years")
plt.ylabel("Temperature (C)")
plt.title("Moving Average Temperature")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data['year'], data['local_avg_temp'], label = 'Actual Local Temp.')
plt.legend()
plt.xlabel("Years")
plt.ylabel("Temperature (C)")
plt.title("Local Average Temperature")
plt.show()
```

```python
plt.figure(figsize=(15, 8))
plt.plot(data['year'], moving_avg(data['local_avg_temp'], 5), label='5 per. Moving Local Temp.',alpha = 0.6)
plt.plot(data['year'], moving_avg(data['local_avg_temp'], 10), label='10 per. Moving Local Temp.',alpha = 0.7)
plt.plot(data['year'], moving_avg(data['local_avg_temp'], 50), label='50 per. Moving Local Temp.')
plt.plot(data['year'], data['local_avg_temp'], label = 'Actual Local Temp.',alpha = 0.4)
plt.legend()
plt.xlabel("Years")
plt.ylabel("Temperature (C)")
plt.title("Moving Average Temperature")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data['year'], data['global_avg_temp'], label = 'Actual Global Temp.',alpha = 0.5)
plt.plot(data['year'], data['local_avg_temp'], label = 'Actual Local Temp.',alpha = 0.7)
plt.legend()
plt.xlabel("Years")
plt.ylabel("Temperature (C)")
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(data['year'], moving_avg(data['global_avg_temp'], 5), label='5 per. Moving Global Temp.')
plt.plot(data['year'], moving_avg(data['global_avg_temp'], 10), label='10 per. Moving Global Temp.')
plt.plot(data['year'], moving_avg(data['local_avg_temp'], 5), label='5 per. Moving Local Temp.')
plt.plot(data['year'], moving_avg(data['local_avg_temp'], 10), label='10 per. Moving Local Temp.')
plt.legend()
plt.xlabel("Years")
plt.ylabel("Temperature (C)")
plt.show()
```
