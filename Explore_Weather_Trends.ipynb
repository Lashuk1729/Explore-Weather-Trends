```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('results.csv')
data.info()
```

```python
data.mean()
data['local_avg_temp'].fillna((data['local_avg_temp'].mean()), inplace=True)
data.info()
```

```python
data.describe()
```

```python
sns.boxplot(data['global_avg_temp'], color = 'orange')
sns.swarmplot(data['global_avg_temp'], color = '.35')
```

```python
sns.boxplot(data['local_avg_temp'], color = 'yellow')
sns.swarmplot(data['local_avg_temp'], color = '.35')
```

```python
def moving_avg(data_series,win):
    out = data_series.rolling(window = win, center = False).mean()
    return out
```

```python
plt.figure(figsize=(15, 6))
plt.plot(data['year'], data['global_avg_temp'], label = 'Actual Global Temp.')
plt.legend()
plt.xlabel("Years")
plt.ylabel("Temperature (C)")
plt.title("Global Average Temperature")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data['year'], data['local_avg_temp'], label = 'Actual Local Temp.')
plt.legend()
plt.xlabel("Years")
plt.ylabel("Temperature (C)")
plt.title("Local Average Temperature")
plt.show()
```
